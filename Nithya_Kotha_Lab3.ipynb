### Task 11
Create a new column titled `CS5000 Grade` in the score DataFrame to calcualte the letter grade for each student in course CS 5000 using the following criteria
+ `A: score >= 90`
+ `B: 80 <= score < 90`
+ `C: 70 <= score < 80`
+ `D: 60 <= score < 70`
+ `F: score < 60`

```python
def function(x):
    
    if x>=90:
        return 'A'
    elif x<90 and x>=80:
        return 'B'
    elif x<80 and x>=70:
        return 'C'
    elif x<70 and x>=60:
        return 'D'
    elif x<60:
        return 'F'
Scores["CS 5000 Grade"]=Scores["CS 5000"].apply(lambda x:function(x))
Scores
```

### Task 13
Using three courses' scores (assuming each course has 3 credits), calculate each student's GPA for Spring 2022, respectively. The following criteria should be used in the calculation
+ `A: 4.0`
+ `B: 3.0`
+ `C: 2.0`
+ `D: 1.0`
+ `F: 0.0`

For example, if a student's grades are A, B and C. That student's GPA should be calculated as

`(3 * 4 + 3 * 3 + 3 * 2) / 9 = 3.0`

Add a new column titled `GPA` in the scores DataFrame to store the GPA for each student

```python
def function(x):
    
    if x>=90:
        return 'A'
    elif x<90 and x>=80:
        return 'B'
    elif x<80 and x>=70:
        return 'C'
    elif x<70 and x>=60:
        return 'D'
    elif x<60:
        return 'F'
Scores_cs_5110=Scores["CS 5110"].apply(lambda x:function(x))
Scores_cs_5600=Scores["CS 5600"].apply(lambda x:function(x))

def Grade(x):
    
    if x =='A':
        return 4
    elif x =='B':
        return 3
    elif x =='C':
        return 2
    elif x =='D':
        return 1
    elif x =='F':
        return 0
Scores_cs_5000_GPA=Scores["CS 5000 Grade"].apply(lambda x:Grade(x))    
Scores_cs_5110_GPA=Scores_cs_5110.apply(lambda x:Grade(x))
Scores_cs_5660_GPA=Scores_cs_5600.apply(lambda x:Grade(x))
Scores['GPA']=(Scores_cs_5000_GPA+Scores_cs_5110_GPA+Scores_cs_5660_GPA)/3
Scores
```

### Task 14
Insert a new column title `Status` to provide a student academic standing based on the GPA
+ `Honor Roll: GPA >= 3.5`
+ `Normal: 3.0 <= GPA < 3.5`
+ `In Probation: GPA < 3.0`

```python
def Statusfunction(x):
    if x>=3.5:
        return "Honor Roll"
    elif x<3.5 and x>=3:
        return "Normal"
    elif x<3:
        return "In Probation"
Scores["Status"]=Scores["GPA"].apply(Statusfunction)
Scores
```
